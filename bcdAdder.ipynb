```python
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute
```

```python
qc=QuantumCircuit(25,5)
```

```python
a = input("Enter a number: ")
b = input("Enter a number: ")
for i, digit in enumerate(a):
    if digit == "1":
        if i == 0:
            qc.x(10)
        if i == 1:
            qc.x(7)
        if i == 2:
            qc.x(4)
        if i == 3:
            qc.x(i)
        
for i, digit in enumerate(b):
    if digit == "1":
        if i == 0:
            qc.x(i)
        if i == 1:
            qc.x(i)
        if i == 2:
            qc.x(i)
        if i == 3:
            qc.x(i)
```

```python

# 1000 + 1001



qc.x(0)
qc.x(10)
qc.x(11)



qc.ccx(0,1,3)
qc.cx(0,1)
qc.ccx(1,2,3)
qc.cx(1,2)


qc.ccx(4,5,6)
qc.cx(4,5)
qc.ccx(3,5,6)
qc.cx(5,3)


qc.ccx(7,8,9)
qc.cx(7,8)
qc.ccx(6,8,9)
qc.cx(8,6)


qc.ccx(10,11,12)
qc.cx(10,11)
qc.ccx(9,11,12)
qc.cx(11,9)


#qc.barrier()



qc.ccx(6,9,13)
qc.ccx(3,9,14)

qc.ccx(12,13,15)
qc.cx(12,13)
qc.cx(13,15)
qc.ccx(14,15,17)
qc.ccx(14,15,16)
qc.cx(15,14)
qc.cx(14,16)
qc.cx(14,17)

#qc.barrier()

qc.measure(17,4)

qc.ccx(18,2,21)
qc.cx(2,18)
qc.ccx(18,20,21)
qc.cx(18,20)
qc.measure(20,0)

qc.ccx(16,3,22)
qc.cx(3,16)
qc.ccx(16,21,22)
qc.cx(16,21)
qc.measure(21,1)

qc.ccx(17,6,23)
qc.cx(6,17)
qc.ccx(17,22,23)
qc.cx(17,22)
qc.measure(22,2)

qc.ccx(19,9,24)
qc.cx(9,19)
qc.ccx(19,23,24)
qc.cx(19,23)
qc.measure(23,3)

qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1024, memory=True)
output = job.result().get_memory()[0]
```
